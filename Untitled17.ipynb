```python
import numpy as np; # have to import numpy as np
a= np.arange(1,10). reshape(3,3) # have to insert number from 1,10 then shape them in matrix

print(a)
```

```python
a[1:,1:2] # 1=[4,5,6] as row and 1:2 = [[5],[8]] as column where [5] is on 1st index position and [8] is on 2 index position.
```

```python
a[1:,2:3] # 1=[4,5,6] as row and 2:3 = [[6],[9]] as column where [6] is on 2nd index position and [9] is on 3rd index position of 2nd row of same column
```

```python
a[0:,1:3] # 0=[1,2,3] as row and 1:3 = [2,3],[5,6],[8,9] as column where [2,3] is on 1st row and [5,6] is on 2nd row and [8,9] is on 3rd colum.
```

```python
a[2:,1:2] # 2=[7,8,9] as row and 1:2 = [[8],[9]] as column where [9] is on 1st index position and [9] is on 2nd index position of 2nd row of same column
```

```python
a[1:,0:1] # 1=[4,5,6] as row and 0:1 = [[4],[7]] as column where 0 is index and  number is of 4 and 1 is pointing 2nd index position of 0 column.
```

```python
a[2:,2:3] # 2=[7,8,9] as row and 2:3 = [9] as column where 2 is index and  number is of 9 and 3 is also at 2nd index position of 2 column.
```
